Contudo, é válido de perguntar, se, por meio de combinação linear, podemos obter todos os vetores de $V$,
em outras palavras, $B$ é uma base para $V = \mathbb{R}^2$?

Para isso, temos que verificar se:

1. Os vetores de $B$ são linearmente independentes
2. Os vetores de $B$ geram $\mathbb{R}^2$

1).Verificaremos a independência linear de $B$:

Se os vetores de $B$ são LI, então:

$$
\sum_{i=1}^{n} \alpha_i \mathbf{v}_i = \mathbf{0} \implies \alpha_1 = \alpha_2 = \dots = \alpha_n = 0
$$

Temos que:
$$  
\alpha (0,1) + \beta (1,0) = (0,0)
$$
\
Montando um sistema de equações:
$$
\begin{cases}
\ \beta=0 \\
\alpha    = 0
\end{cases}
$$
\
Resolvendo o sistemas de equações, temos que:
$$\alpha,\beta=0$$
\
Logo, os vetores $B$ são LI



2). Mostrar que consigo gerar $V$


Para isso, precisamos mostrar que caso eu faça uma combinação linear com os vetores da base tentando alcançar vetores quaisquer $(x,y) \in V$ existem escalares que satisfazem essa proposição.

Primariamente, assumimos que:

\
$$\alpha (0,1) + \beta (1,0) = (x,y); \ x,y \in
\mathbb{R} $$
\
Dessa forma, obtemos um sistema de equações:
$$
\begin{cases}
\ \beta=x \\
\alpha  = y
\end{cases}
$$

Resolvendo o sistemas de equações, temos que:
$$\alpha=y;\beta=x$$

Portanto, $B$ gera $V$. Deste modo, $B$ é base de $V$


Definimos duas transformações lineares $\
T1: \mathbb{R}^2   \to \mathbb{R}^2$ e $\
T2: \mathbb{R}^2 \to \mathbb{R}^2$ com as seguintes caracteristicas:

\
$$T1(x,y)=(-y,x)  $$
$$T2(x,y)=(x+2y,y) $$
\
$T1$ é uma transformação linear que recebe como entrada um vetor que rotaciona o vetor de entrada em 90º. $T2$ é uma transformação linear que cisalha x proporcionalmente ao dobro de y.

Visualmente, temos que:


### **Transformação T1: Rotação em 90º**

```python
from matplotlib import pyplot as plt

#Metainformação
plt.grid(True)
plt.title("Rotação em 90º")
plt.axhline(0, color='black', linewidth=1)  # eixo X
plt.axvline(0, color='black', linewidth=1)  # eixo Y

#Escala
plt.xlim(-10,10)
plt.ylim(-10,10)

#Entrada do vetor
x=int(input("informe o componente x de seu vetor: "))
y=int(input("informe o componente y de seu vetor: "))

def vetor(x,y):
   plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='green')
   return [x,y]

def T1(vetor):
  xt=-1*vetor[1]
  yt=vetor[0]
  plt.quiver(0, 0, xt, yt, angles='xy', scale_units='xy', scale=1, color='red')
  return [xt,yt]

vetorinput=vetor(x,y)
print("seu vetor de entrada é:",vetorinput)
print("seu vetor transformado é:",T1(vetorinput))
plt.show()
```

# Nova seção

Além da representação geometrica, podemos representar um vetor utilizando matrizes. Tomando como exemplo as bases canônicas que definimos, temos:

$$u=\begin{pmatrix}
1   \\
0
\end{pmatrix} \ \ \ \ \ \ \ \ \ v= \begin{pmatrix}
0   \\
1
\end{pmatrix}$$
\
Por conveniência, poderiamos organizá-los em uma matriz de bases:
$$B=\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
$$

Para obter matriz de transformação, iremos:
 1. aplicar a transformação linear em cada vetor da base utilizando os algoritmos acima  
 2. organizar os valores na matriz como no exemplo acima:
 3. mudar o sistema de coordenadas da matriz de base
\
\
Matriz de rotação $T1$:


```python
from matplotlib import pyplot as plt

#Metainformação
plt.grid(True)
plt.title("Rotacionar em 90°")
plt.axhline(0, color='black', linewidth=1)  # eixo X
plt.axvline(0, color='black', linewidth=1)  # eixo Y

#Escala
plt.xlim(-10,10)
plt.ylim(-10,10)

#Entrada do vetor



def vetor(x,y):
   plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='green')
   return [x,y]

def criarvetor():
    x=int(input("informe a componente x de seu vetor: "))
    y=int(input("informe a componente y de seu vetor: "))
    vetorinput=vetor(x,y)
    return vetorinput

def T1(vetor):
  xt=-1*vetor[1]
  yt=vetor[0]
  return [xt,yt]

def matriz_vetores(base1, base2):
    vetor1transformado = T1(base1)
    vetor2transformado = T1(base2)

    # matriz transformada
    matriz_transformacao = np.array([
        [vetor1transformado[0], vetor2transformado[0]],
        [vetor1transformado[1], vetor2transformado[1]]
    ])

    print("Valor da matriz transformada com as coordenadas antigas : ", matriz_transformacao)

    # Base canônica
    P = np.array([[1, 0],
                  [0, 1]])
    P_inv = np.linalg.inv(P)

    # Matriz na nova base
    matriz_resultado = P_inv @ matriz_transformacao @ P

    # Plot
    plt.quiver(0, 0, matriz_resultado[0, 0], matriz_resultado[1, 0], angles='xy', scale_units='xy', scale=1, color='red')
    plt.quiver(0, 0, matriz_resultado[0, 1], matriz_resultado[1, 1], angles='xy', scale_units='xy', scale=1, color='red')

    print("Valor da matriz transformada com as coordenadas novas : ", matriz_resultado)
    return matriz_resultado


vetor1 = criarvetor()
vetor2 = criarvetor()
matriz_vetores(vetor1, vetor2)



plt.show()
```

# Transformação linear a partir da matriz  

A partir das matrizes transformada fornecida, podemos utilizar operações fundamentais de produto e soma de matrizes para realizar transformações lineares em vetores genericos $ (x,y) \in V. $ Deste modo, temos:

```python
from matplotlib import pyplot as plt
import numpy as np

#Metainformação
plt.grid(True)
plt.title("Rotacionar em 90° por matriz")
plt.axhline(0, color='black', linewidth=1)  # eixo X
plt.axvline(0, color='black', linewidth=1)  # eixo Y

#Escala
plt.xlim(-10,10)
plt.ylim(-10,10)

#Entrada do vetor



def vetor(x,y):
   plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='green')
   return [x,y]

def criarvetor():
    x=int(input("informe a componente x de seu vetor: "))
    y=int(input("informe a componente y de seu vetor: "))
    vetorinput=vetor(x,y)
    return vetorinput


def transformacao_matriz(vetorarbitrario):
    matrizresultado=np.array([0,0])
    matrizbasetransformada=np.array([[0,1],[-1,0]])
    vetorarbitrario=np.array(vetorarbitrario)
    matrizresultado=matrizbasetransformada @ vetorarbitrario
    plt.quiver(0, 0, matrizresultado[0], matrizresultado[1], angles='xy', scale_units='xy', scale=1, color='red')
    print( matrizresultado)
    return matrizresultado



vetor1=criarvetor()
vetor2=criarvetor()
transformacao_matriz(vetor1)
transformacao_matriz(vetor2)


plt.show()
```
